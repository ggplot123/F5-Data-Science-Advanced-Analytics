```python
import theano as T
import theano.tensor as TT
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 200
a = 4
b = 2

points_x = np.linspace(0,10,N)
points_y = a+np.sin(b*points_x) + np.random.RandomState(1).normal(0,0.3,N)
data = np.array([points_x,points_y]).T
#Generate a noisy sine curve for data.
#Note that this is a nonlinear function so OLS isn't going to work even with transformations.
#Well, I guess you could do a cubic regression but it really isn't going to work outside [0,10]
plt.scatter(points_x, points_y)
```

```python
data_placeholder = TT.matrix()
a_var = TT.scalar()
b_var = TT.scalar()
cost = T.function(inputs = [data_placeholder, a_var, b_var], 
               outputs = TT.sum(((a_var + TT.sin(data_placeholder[:,0]*b_var)) - data_placeholder[:,1])**2))
c =  TT.sqrt(TT.sum(((a_var + TT.sin(data_placeholder[:,0]*b_var)) - data_placeholder[:,1])**2))
```

```python
d_grad = T.function([data_placeholder, a_var, b_var], TT.sqrt(TT.sum(TT.pow(T.grad(c, [a_var, b_var]), 2))))
```

```python
z = np.vectorize(lambda x,y:d_grad(data, x, y)[()],  otypes = [np.float64])
#z = np.vectorize(lambda x,y: d_grad(data, x, y),
```

```python
xx, yy = np.meshgrid(np.linspace(1,6), np.linspace(0,2))
```

```python
zz = z(xx, yy)
```

```python
plt.figure(figsize=(10,10))
CS = plt.contourf(xx,yy,zz, levels = np.linspace(0,64,200), cmap = 'terrain')
plt.colorbar()
```
